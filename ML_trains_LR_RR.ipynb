```python
# Imports
import pandas as pd
import numpy as np
from loss import ape
```

```python
# Importing dataset
folder_location = 'data'
file_name = 'trains_dataset.csv'
df = pd.read_csv(open(folder_location + '\\' + file_name))
```

```python
df
```

```python
def train_test_split(df, no_of_years_in_test_case):
    xtrain = pd.DataFrame(columns = df.columns[:-1])
    ytrain = pd.DataFrame(columns = [df.columns[-1]])
    xtest = pd.DataFrame(columns = df.columns[:-1])
    ytest = pd.DataFrame(columns = [df.columns[-1]])
    
    s_id = df['station_id'].unique()
    
    for s in s_id:
        a = df.loc[df['station_id'] == s]
        xtrain = xtrain.append(a.iloc[:-no_of_years_in_test_case, :-1])
        ytrain = ytrain.append(a.iloc[:-no_of_years_in_test_case, -1:])
        xtest = xtest.append(a.iloc[-no_of_years_in_test_case:, :-1])
        ytest = ytest.append(a.iloc[-no_of_years_in_test_case:, -1:])

    return xtrain, ytrain, xtest, ytest
```

```python
xtrain, ytrain, xtest, ytest = train_test_split(df, 1)
```

```python
ytest
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge

lr = LinearRegression()
lr.fit(xtrain, ytrain)

rr = Ridge(alpha=0.01)
rr.fit(xtrain, ytrain)

rr100 = Ridge(alpha=100)
rr100.fit(xtrain, ytrain)

y_pred_lr = lr.predict(xtest)
y_pred_rr = rr.predict(xtest)
y_pred_rr100 = rr100.predict(xtest)
```

```python
loss_lr = ape(ytest.values, y_pred_lr)
loss_rr = ape(ytest.values, y_pred_rr)
loss_rr100 = ape(ytest.values, y_pred_rr100)

print('Absolute Percentage Loss (%)')
print('Linear Regression Loss: ' + str(round(loss_lr[0], 2)))
print('Ridge Regression Loss: ' + str(round(loss_rr[0], 2)))
print('Ridge Regression (with alpha 100) Loss: ' + str(round(loss_rr100[0], 2)))
```
